```python
# writing data to csv file
# to_csv is method of DataFrame <-- give name of csv file (created if not availabe)
# by default it also writes index (1,2,3...). to avoid this: set index = False
# to write any specific columns i.e not all columns: columns = [columns name]
df5.to_csv("new.csv", index = False, columns =["Country", "Item Type"])
```

```python
# now for Excel file
dfx = pd.read_excel("data/sample_data.xls")
```

```python
# we can use converter functions in excel DF
def sample_converter(cell):
    if cell == "n.a" or cell == "H":
        return "NaN"
    return cell
# the converter is called at each cell, and values are converted accordingly
dfx = pd.read_excel("data/sample_data.xls",nrows = 5,converters = {
    "Order Priority":sample_converter
})
dfx
```

```python
# write to excel file
# sheet name is optional
# startrow and startcol are used to provide offset for writing in xls file
dfx.to_excel("new.xls",sheet_name ="Sheet1", index = False, startrow=5,startcol=2)
```

```python
# writing to two (or more) sheets in a xls file
# using ExcelWriter class
with pd.ExcelWriter("new.xls") as writer:
    dfx.to_excel(writer, sheet_name="sheet1")
    dfx.to_excel(writer,sheet_name = "sheet2")
```
